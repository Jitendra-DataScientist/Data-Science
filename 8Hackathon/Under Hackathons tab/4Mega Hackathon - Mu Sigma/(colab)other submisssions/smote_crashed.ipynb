```python
df.drop('ID',axis=1,inplace=True)
```

```python
df['Postal Code'] = df['Postal Code'].astype("object")
```

```python
df['Loan/No Loan'].value_counts()
```

```python
df['Loan/No Loan'].value_counts(normalize=True)
```

```python
df['Yearly Income'].fillna(df['Yearly Income'].mean(),inplace=True)
df['Debt to Income'].fillna(df['Debt to Income'].mean(),inplace=True)
df['Total Unpaid CL'].fillna(df['Total Unpaid CL'].mean(),inplace=True)
df['Unpaid Amount'].fillna(df['Unpaid Amount'].mean(),inplace=True)

df['Postal Code'].fillna(df['Postal Code'].value_counts().index[0],inplace=True)
df['Designation'].fillna(df['Designation'].value_counts().index[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.info()
```

```python
X = df.drop('Loan/No Loan', axis=1)

y = df['Loan/No Loan']
```

```python
categorical_features_indices = np.where(X.dtypes == "object")[0]
categorical_features_indices
```

```python
from imblearn.over_sampling import SMOTENC
smote_nc = SMOTENC(categorical_features=categorical_features_indices, random_state=0)
X_resampled, y_resampled = smote_nc.fit_resample(X, y)
```

```python
X = X_resampled.copy()
y_resampled
```

```python
pip install catboost
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier(random_state=1,cat_features=categorical_features_indices,eval_metric='F1',
            learning_rate=0.32)
model=clf.fit(X,y)
```
