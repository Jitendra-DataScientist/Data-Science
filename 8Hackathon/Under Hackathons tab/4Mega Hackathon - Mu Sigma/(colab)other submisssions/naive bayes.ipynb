```python
numerical_features_indices = np.where(df.dtypes != "object")[0]

lst=[]
for i in range(len(df.columns)):
  if i in numerical_features_indices:
    #print(df.columns[i])
    lst.append(df.columns[i])

df_temp = df[lst]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df_temp)
df_temp=pd.DataFrame(scaled,columns=df_temp.columns,index=df_temp.index)

lst1=[]
for i in range(len(df.columns)):
  if i not in numerical_features_indices:
    print(df.columns[i])
    lst1.append(df.columns[i])

df_temp1= df[lst1]

df1 = pd.concat([df_temp, df_temp1], axis=1)
df = df1.copy()
```
