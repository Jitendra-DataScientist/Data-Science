```python
df['Experience'].replace('4yrs','4-9',inplace=True)
df['Experience'].replace('5yrs','4-9',inplace=True)
df['Experience'].replace('6yrs','4-9',inplace=True)
df['Experience'].replace('7yrs','4-9',inplace=True)
df['Experience'].replace('8yrs','4-9',inplace=True)
df['Experience'].replace('9yrs','4-9',inplace=True)
```

```python
df['Unpaid 2 years'].value_counts(normalize=True)
```

```python
for i in range(df.shape[0]):
    if (df.loc[i,'Unpaid 2 years'] != '0') and (df.loc[i,'Unpaid 2 years'] != '1') and (df.loc[i,'Unpaid 2 years'] != '2'):
        df.loc[i,'Unpaid 2 years'] = 3
```

```python
df['Already Defaulted'] = df['Already Defaulted'].astype(str)
df.info()
```

```python
df['Already Defaulted'].value_counts(normalize=True)
```

```python
for i in range(df.shape[0]):
    if df.loc[i,'Already Defaulted'] != '0':
        df.loc[i,'Already Defaulted'] = '1'
```

```python
df['Designation'].value_counts(normalize=True)
```

```python
df['Postal Code'] = df['Postal Code'].astype(str)
df['Postal Code'].value_counts(normalize=True)
```

```python
df['Deprecatory Records'] = df['Deprecatory Records'].astype(str)
df['Deprecatory Records'].value_counts(normalize=True)
```

```python
for i in range(df.shape[0]):
    if (df.loc[i,'Deprecatory Records'] != '0') and (df.loc[i,'Deprecatory Records'] != '1'):
        df.loc[i,'Deprecatory Records'] = '2'
```

```python
df['Inquiries'] = df['Inquiries'].astype(str)
df['Inquiries'].value_counts(normalize=True)
```

```python
for i in range(df.shape[0]):
    if (df.loc[i,'Inquiries'] != '0') and (df.loc[i,'Inquiries'] != '1') and (df.loc[i,'Inquiries'] != '2'):
        df.loc[i,'Inquiries'] = '3'
```

```python
df['Reason'].value_counts(normalize=True)
```

```python
for i in range(df.shape[0]):
    if (df.loc[i,'Inquiries'] != 'debt  consolidation') and (df.loc[i,'Inquiries'] != 'credit  card') and (df.loc[i,'Inquiries'] != 'home  improvement') and (df.loc[i,'Inquiries'] != 'other'):
        df.loc[i,'Inquiries'] = 'other-1'
```

```python
df.drop(['Designation','Asst_Reg','Postal Code','Claim Type'],axis=1,inplace=True)
```

```python
df.drop(['Account Open', 'State', 'Sub GGGrade'],axis=1,inplace=True)
```

```python
df.drop('ID',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Loan/No Loan'].value_counts()
```

```python
df['Loan/No Loan'].value_counts(normalize=True)
```

```python
df['Yearly Income'].fillna(df['Yearly Income'].mean(),inplace=True)
df['Debt to Income'].fillna(df['Debt to Income'].mean(),inplace=True)
df['Total Unpaid CL'].fillna(df['Total Unpaid CL'].mean(),inplace=True)
df['Unpaid Amount'].fillna(df['Unpaid Amount'].mean(),inplace=True)

#df['Postal Code'].fillna(df['Postal Code'].value_counts().index[0],inplace=True)
#df['Designation'].fillna(df['Designation'].value_counts().index[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.info()
```

```python
X = df.drop('Loan/No Loan', axis=1)

y = df['Loan/No Loan']
```

```python
categorical_features_indices = np.where(X.dtypes == "object")[0]
categorical_features_indices
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier(random_state=1,cat_features=categorical_features_indices,eval_metric="F1",iterations=3000)
model=clf.fit(X,y)
```
