```python
from sklearn.metrics import f1_score
import numpy as np

def f1_eval(y_pred, dtrain):
    y_true = dtrain.get_label()
    err = 1-f1_score(y_true, np.round(y_pred))
    return 'f1_err', err
```

```python
from xgboost import XGBClassifier
clf = XGBClassifier(tree_method='gpu_hist',sampling_method='gradient_based',subsample=0.2,iterations=20000,eta=0.01,gamma = 1000, updater='grow_gpu_hist')
# clf = XGBClassifier(tree_method='gpu_hist',sampling_method='gradient_based',subsample=0.2,iterations=10000,eta=0.1,gamma = 1000,updater='refresh')
# clf = XGBClassifier(tree_method='gpu_hist',sampling_method='gradient_based',subsample=0.2,iterations=10000,eta=0.1,booster='gblinear',reg_lambda=5.0 )
model=clf.fit(X,y,eval_metric=f1_eval)
```

```python
df['Loan/No Loan'] = model.predict(df.drop('ID',axis=1))
df
```

```python
pred_prob = model.predict_proba(df.drop(['ID','Loan/No Loan'],axis=1))
pred_prob = pred_prob[:,1]
pred_prob
```

```python
custom_prob = 0.24
custom_cutoff_data=[]

for i in range(0,len(pred_prob)):  #2333
    if pred_prob[i] >custom_prob:
        a= 1
    else:
        a=0
    custom_cutoff_data.append(a)
```

```python
df['Loan/No Loan'] = custom_cutoff_data
```
