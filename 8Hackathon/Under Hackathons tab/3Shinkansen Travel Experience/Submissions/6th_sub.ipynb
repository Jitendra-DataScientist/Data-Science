```python
df['Seat_comfort'].fillna(df['Seat_comfort'].value_counts().index[0],inplace=True)
df['Arrival_time_convenient'].fillna(df['Arrival_time_convenient'].value_counts().index[0],inplace=True)
df['Catering'].fillna(df['Catering'].value_counts().index[0],inplace=True)
df['Platform_location'].fillna(df['Platform_location'].value_counts().index[0],inplace=True)
df['Onboardwifi_service'].fillna(df['Onboardwifi_service'].value_counts().index[0],inplace=True)
df['Onboard_entertainment'].fillna(df['Onboard_entertainment'].value_counts().index[0],inplace=True)
df['Online_support'].fillna(df['Online_support'].value_counts().index[0],inplace=True)
df['Onlinebooking_Ease'].fillna(df['Onlinebooking_Ease'].value_counts().index[0],inplace=True)
df['Onboard_service'].fillna(df['Onboard_service'].value_counts().index[0],inplace=True)
df['Leg_room'].fillna(df['Leg_room'].value_counts().index[0],inplace=True)
df['Baggage_handling'].fillna(df['Baggage_handling'].value_counts().index[0],inplace=True)
df['Checkin_service'].fillna(df['Checkin_service'].value_counts().index[0],inplace=True)

df['Cleanliness'].fillna(df['Cleanliness'].value_counts().index[0],inplace=True)
df['Online_boarding'].fillna(df['Online_boarding'].value_counts().index[0],inplace=True)
df['Gender'].fillna(df['Gender'].value_counts().index[0],inplace=True)
df['CustomerType'].fillna(df['CustomerType'].value_counts().index[0],inplace=True)
df['TypeTravel'].fillna(df['TypeTravel'].value_counts().index[0],inplace=True)
```

```python
df.select_dtypes("int").isnull().sum()
```

```python
df.select_dtypes("float").isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['DepartureDelay_in_Mins'].fillna(df['DepartureDelay_in_Mins'].mean(),inplace=True)
df['ArrivalDelay_in_Mins'].fillna(df['ArrivalDelay_in_Mins'].mean(),inplace=True)
```

```python
for column in df.columns: 
    if df[column].dtype == 'object':
        print(column.upper(),': ',df[column].nunique(),'unique values')
        print('--------------------')
        print(df[column].value_counts())
        print('\n',column.upper(),'(normalized)')
        print('--------------------')
        print(df[column].value_counts(normalize=1))
        print('\n*************************************************\n\n')
```

```python
#df = pd.get_dummies(df,drop_first=True)
X = df.drop(['Overall_Experience','ID'], axis=1)

y = df['Overall_Experience']
X
```

```python
categorical_features_indices = np.where(X.dtypes != np.float)[0]
categorical_features_indices
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier(random_state=1,cat_features=categorical_features_indices,task_type='CPU',max_depth=9,l2_leaf_reg=2.0,n_estimators=3000)   
model=clf.fit(X,y)
```
