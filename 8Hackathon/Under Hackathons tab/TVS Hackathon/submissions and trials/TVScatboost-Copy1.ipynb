```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Train_Set_TVS.csv')
df
```

```python
df.drop('ID',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Compensation'].fillna(df['Compensation'].mean(),inplace=True)
df['Annuity_amt'].fillna(df['Annuity_amt'].mean(),inplace=True)
df['density_Ind'].fillna(df['density_Ind'].mean(),inplace=True)
df['Home_Appliances'].fillna(df['Home_Appliances'].mean(),inplace=True)
df['Experience'].fillna(df['Experience'].mean(),inplace=True)
df['Company_Rating'].fillna(df['Company_Rating'].mean(),inplace=True)
df['c1_rated'].fillna(df['c1_rated'].mean(),inplace=True)
df['c2_rated'].fillna(df['c2_rated'].mean(),inplace=True)
df['Contact_age'].fillna(df['Contact_age'].mean(),inplace=True)


df['Vehicle_Status'].fillna(df['Vehicle_Status'].value_counts().index[0],inplace=True)
df['Job_Type'].fillna(df['Job_Type'].value_counts().index[0],inplace=True)
df['Prop_List'].fillna(df['Prop_List'].value_counts().index[0],inplace=True)
df['Credit_Inq'].fillna(df['Credit_Inq'].value_counts().index[0],inplace=True)
df['Income_Alt'].fillna(df['Income_Alt'].value_counts().index[0],inplace=True)
```

```python
df.dropna(inplace=True)
df
```

```python
df.info()
```

```python
# df = pd.get_dummies(df,drop_first=True)
X = df.drop('Default', axis=1)

y = df['Default']
```

```python
X.drop(['Company_Rating','Age','Annuity_amt','Dependents','Default_hist','Home_Appliances','c1_rated','hour_Process','Marital_Stat','Addr_permanent'],axis=1,inplace=True)
```

```python
categorical_features_indices = np.where(X.dtypes == "object")[0]
categorical_features_indices
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier(random_state=1,cat_features=categorical_features_indices,eval_metric='F1',learning_rate=0.3)
model=clf.fit(X,y)
```
