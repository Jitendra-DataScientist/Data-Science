```python
#How to change the font of x and y axis? How to resize the x and y axis labels?  How to name a plot?

import matplotlib.pyplot as plt
  
# setting font sizeto 30
plt.rcParams.update({'font.size': 30})
  
x = [1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10]
  
# plotting a plot
plt.plot(x, y)
  
# setting title name
plt.title("Title")
  
# setting x axis label
plt.xlabel("x axis")
  
# setting y axis label
plt.ylabel("y axis")
  
plt.show()
```

```python
#How to resize and change font of Hue?
import seaborn as sns
import matplotlib.pylab as plt
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")

from pylab import rcParams
rcParams['figure.figsize'] = 15, 10


sns.stripplot(x="sex", y="total_bill", hue="day", data=tips, jitter=True) #, figsize = (15,5))
#plt.setp(ax.get_legend().get_texts(), fontsize='22') # for legend text
#plt.setp(ax.get_legend().get_title(), fontsize='32') # for legend title
plt.legend(fontsize='x-large', title_fontsize='40')


plt.show()
```
