```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
#reading file
auto=pd.read_csv('Automobile (1).csv')
auto.head()
```

```python
#plotting univariate distributions
sns.distplot(auto['highway_mpg']); #without 
```

```python
sns.distplot(auto['city_mpg'],kde=False,rug=True);
```

```python
#plotting bivariate variables
sns.jointplot(x=auto['engine_size'], y=auto['horsepower']);
```

```python
#hex plot
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'],kind='hex');
```

```python
#kernel density estimation
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'],kind='kde');
```

```python
#pairplot()
sns.pairplot(auto[['wheel_base','engine_size','bore']]);
```

```python
#categorical plotting
sns.stripplot(y=auto['horsepower'],x=auto['fuel_type'],jitter=True)
```

```python
#swarm plot
sns.swarmplot(x=auto['fuel_type'],y=auto['horsepower']);
```

```python
#additional content for matplotlib
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#lineplot
x=np.arange(1,11)
y=np.arange(2,21,2)
plt.plot(x,y)
#plt.show()
```

```python
#adding labels and title
x=np.arange(1,11)
y=x*2
plt.plot(x,y)
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#changing line aesthetics
x=np.arange(1,11)
y=x*2
plt.plot(x,y,color='r',linestyle=':',linewidth=6)
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
multiple plots
x=np.arange(1,11)
y=x*2
z=x*3
plt.plot(x,y,color='r',linestyle=':',linewidth=2)
plt.plot(x,z,color='g',linestyle='-',linewidth=4)
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(False)
plt.show()
```

```python
#sub plots
x=np.arange(1,11)
y=x*2
z=x*3
plt.subplot(2,1,1)
plt.plot(x,y,color='r',linestyle='-.',linewidth=2)
plt.title('Line plot 1')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(False)
plt.subplot(2,1,2)
plt.plot(x,z,color='g',linestyle='-',linewidth=4)
plt.title('Line plot 2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(False)
plt.show()
```

```python
#scatter plot
x=[10,20,30,40,50,50,70,80,90]
y=[4,6,5,1,3,2,7,9,9]
plt.scatter(x,y)
plt.show()
```

```python
#changing marker aesthetics
x=[10,20,30,40,50,50,70,80,90]
y=[4,6,5,1,3,2,7,9,9]
plt.scatter(x,y,color='g',marker='*',s=500)
plt.show()
```

```python
#multiple plots
x=[10,20,30,40,50,50,70,80,90]
y=[4,6,5,1,3,2,7,9,9]
z=[6,4,7,5,3,1,2,9,8]
plt.scatter(x,y,color='g',marker='*',s=200)
plt.scatter(x,z,color='violet',marker='*',s=400)
plt.show()
```

```python
#subplots
#multiple plots
x=[10,20,30,40,50,50,70,80,90]
y=[4,6,5,1,3,2,7,9,9]
z=[6,4,7,5,3,1,2,9,8]
plt.subplot(1,2,1)
plt.scatter(x,y,color='g',marker='*',s=200)
plt.subplot(1,2,2)
plt.scatter(x,z,color='violet',marker='*',s=400)
plt.show()
```

```python
#bar plot
student={'Bob':87,'Matt':56,"Jin":27}
nam=list(student.keys())
val=list(student.values())
plt.bar(nam,val)
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('Hi all')
plt.show()
```

```python
#horizontal bar plot
student={'Bob':87,'Matt':56,"Jin":27}
nam=list(student.keys())
val=list(student.values())
plt.barh(nam,val,color='coral')
plt.ylabel('Names')
plt.xlabel('Marks')
plt.title('Hi all')
plt.show()
```

```python
#box plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,8,9,7,8,9,6,8,6]
data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

```python
#violin plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,8,9,7,8,9,6,8,6]
data=list([one,two,three])
plt.violinplot(data,showmedians=True,showextrema=False,showmeans=True)
plt.show()
```

```python
#pie chart
fruit=['apple','guava','mango','orange']
quantity=[34,65,14,86]
plt.pie(quantity,labels=fruit,shadow=True,colors=['lavender','coral','burlywood','magenta'],autopct='%0.3f%%')
plt.show()
```

```python
#doughnut chart
fruit=['apple','guava','mango','orange']
quantity=[34,65,14,86]
plt.pie(quantity,labels=fruit,radius=3,shadow=True,colors=['lavender','coral','burlywood','magenta'],autopct='%0.5f%%')
plt.pie([1],colors=['grey'],radius=1)
plt.show()
```
