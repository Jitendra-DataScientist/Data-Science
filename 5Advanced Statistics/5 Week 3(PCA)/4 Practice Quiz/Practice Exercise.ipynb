```python
from scipy.stats import zscore
data_scaled=df.apply(zscore)
data_scaled.drop('Unnamed: 0',axis=1,inplace=True)
data_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=11, random_state=123)   # 'n_components=12' as we want eigen values for all variables, 'random_state=123' as we want consistent output
pca_transformed = pca.fit_transform(data_scaled)
pca_transformed
```
